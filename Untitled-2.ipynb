```python
import tkinter as tk
from tkinter import messagebox
import json

class MedicalExpertGUI:
    def __init__(self, root, kb):
        self.root = root
        self.kb = kb
        self.vars = {}
        root.title("Medical Expert System")

        # Create checkboxes for each symptom
        for i, sym in enumerate(kb["symptoms"]):
            var = tk.BooleanVar()
            chk = tk.Checkbutton(root, text=sym, variable=var)
            chk.grid(row=i//2, column=i%2, sticky="w")
            self.vars[sym] = var

        tk.Button(root, text="Diagnose", command=self.diagnose).grid(row=len(kb["symptoms"])//2+1, column=0, pady=10)
        tk.Button(root, text="Quit", command=root.quit).grid(row=len(kb["symptoms"])//2+1, column=1)

    def diagnose(self):
        user_symptoms = [sym for sym, var in self.vars.items() if var.get()]
        possible = []
        for rule in self.kb["rules"]:
            if all(cond in user_symptoms for cond in rule["condition"]):
                possible.append(rule["conclusion"])
        if possible:
            messagebox.showinfo("Diagnosis", f"Possible illness(es): {', '.join(possible)}")
        else:
            messagebox.showwarning("No match", "No disease matches your symptoms. See a doctor.")

if __name__ == "__main__":
    with open("knowledgebase.json") as f:
        kb = json.load(f)
    root = tk.Tk()
    app = MedicalExpertGUI(root, kb)
    root.mainloop()
```
