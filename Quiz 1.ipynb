```python
class CropDiseaseKB:
    def __init__(self, kb_dict):
        self.facts = kb_dict.get("facts", {})
        self.rules = kb_dict.get("rules", [])

    def get_fact(self, disease):
        return self.facts.get(disease, "Fact unavailable.")

    def diagnose(self, observed_symptoms):
        observed = [sym.strip().lower() for sym in observed_symptoms]
        matches = []

        for rule in self.rules:
            disease = rule.get("disease")
            required_symptoms = rule.get("symptoms") or rule.get("required_symptoms") or []
            required = [sym.strip().lower() for sym in required_symptoms]

            if required and all(sym in observed for sym in required):
                matched = [sym for sym in required if sym in observed]
                matches.append({
                    "disease": disease,
                    "matched_symptoms": matched,
                    "confidence": rule.get("confidence", 1.0),
                })

        return matches

def main():
    print(" Kenyan Crop Disease Diagnosis System \n")
    
    kb = CropDiseaseKB(cropdiseases_kb)
    
    try:
        # If a function exists
        from disease_recommendations import get_recommendations
    except ImportError:
        # In Jupyter, maybe it's just a dict named disease_recommendations in global scope
        if 'disease_recommendations' in globals():
            rec_dict = globals()['disease_recommendations']
            def get_recommendations(disease):
                return rec_dict.get(disease)
        else:
            def get_recommendations(disease):
                return None
    
    # Accept user input
    print("Please enter the symptoms you observe (comma-separated):")
    print("Example: maize, chlorotic mottling, dead heart, leaf drying from edges")
    user_input = input("Symptoms: ").strip().lower()
    observed_symptoms = [sym.strip() for sym in user_input.split(",") if sym.strip()]
    
    if not observed_symptoms:
        print("No symptoms entered. Exiting.")
        return
    
    print(f"\nObserved symptoms: {', '.join(observed_symptoms)}\n")
    print("=== Reasoning Steps ===\n")
    
    # Apply rules (forward chaining)
    matches = kb.diagnose(observed_symptoms)
    
    if not matches:
        print("No matching disease found in the knowledge base.")
        print("Please consult a local agricultural officer for further investigation.")
        return
    
    conclusions = []
    for idx, match in enumerate(matches, 1):
        disease = match["disease"]
        matched_symptoms = match["matched_symptoms"]
        confidence = match["confidence"]
        
        print(f"Step {idx}: Rule triggered for {disease}")
        print(f"   Because symptoms observed: {', '.join(matched_symptoms)}")
        print(f"   → All required symptoms matched → {disease} suspected.")
        
        # Get fact from KB
        fact = kb.get_fact(disease)
        print(f"   Fact: {fact}")
        
        # Get recommendations
        rec = get_recommendations(disease)
        if rec:
            print("   Management recommendations:")
            for category, advice_list in rec.items():
                cat_name = category.replace('_',' ').title()
                print(f"      {cat_name}:")
                for item in advice_list[:2]:
                    print(f"        - {item}")
                if len(advice_list) > 2:
                    print(f"        ... and {len(advice_list)-2} more.")
        else:
            print("   No specific recommendations available.")
        
        print(f"   Confidence: {confidence}\n")
        conclusions.append(disease)
    
    # Multi-step reasoning based on facts
    print("=== Multi‑Step Reasoning ===")
    if len(conclusions) > 1:
        print(f"Multiple diseases detected: {', '.join(conclusions)}")
        # Check for known co-infections
        if "Cassava Brown Streak Disease (CBSD)" in conclusions and "Cassava Mosaic Disease (CMD)" in conclusions:
            print("→ Fact 15: CBSD and CMD often co-infect cassava, increasing severity.")
            print("→ Recommendation: Use certified virus‑free cuttings and remove infected plants immediately.")
        elif "Bean Common Mosaic Virus (BCMV)" in conclusions and len(conclusions)>1:
            print("→ Fact 6: Eight different viruses infect common beans in Western Kenya, often as co‑infections.")
            print("→ Recommendation: Consider lab testing to identify all viral strains before next planting.")
        else:
            print("→ Co‑infection possible. Monitor fields closely and consult an agricultural officer.")
    elif len(conclusions) == 1:
        disease = conclusions[0]
        if disease == "Potato/Tomato Late Blight":
            print("→ Fact 12: 65.8 percent of farmers in Nyandarua use only one fungicide type, leading to resistance.")
            print("→ Multi‑step reasoning: If you have used only metalaxyl before, rotate to mancozeb or cymoxanil.")
        elif disease == "Maize Lethal Necrosis (MLN)":
            print("→ Fact 3: MLN was first detected in Rift Valley (2011) and can cause 100% yield loss.")
            print("→ Multi‑step reasoning: Because MLN is viral and vector‑borne, control aphids/thrips alongside roguing.")
        elif disease == "Wheat Stem Rust":
            print("→ Fact 4: Kenya is a global hotspot for Ug99 race group overcoming many resistant varieties.")
            print("→ Multi‑step reasoning: Even with resistant varieties, monitor for new strains and apply triazoles early.")
        else:
            print(f"→ Background fact: {kb.get_fact(disease)}")
            print("→ Multi‑step reasoning: Prioritize preventive measures for next season based on this fact.")
    else:
        print("No conclusions to reason further.")
    
    print("\n=== Final Conclusion ===")
    if len(conclusions) == 1:
        print(f"The most likely disease is: {conclusions[0]}")
    else:
        print(f"Possible diseases: {', '.join(conclusions)}")
        print("Consider professional field diagnosis to confirm co‑infection.")
    
    print("\nThank you for using the Kenyan Crop Disease Diagnosis System.")

# Run the main function
if __name__ == "__main__":
    main()
```
