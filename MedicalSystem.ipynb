```python
import json

def load_knowledge_base(filepath="knowledgebase.json"):
    with open(filepath, 'r') as f:
        return json.load(f)

def get_user_symptoms(symptoms):
    #Ask user about each symptom with validation.
    user_symptoms = []
    print("\nPlease answer the following questions (y/n):")
    for sym in symptoms:
        while True:
            answer = input(f"Do you have {sym}? ").strip().lower()
            if answer in ['y', 'yes']:
                user_symptoms.append(sym)
                break
            elif answer in ['n', 'no']:
                break
            else:
                print("Invalid input. Please enter 'y' or 'n'.")
    return user_symptoms

def infer_diseases(rules, user_symptoms):
    #Return list of diseases whose conditions are all satisfied.
    possible = []
    for rule in rules:
        condition = rule["condition"]
        if all(symptom in user_symptoms for symptom in condition):
            possible.append(rule["conclusion"])
    return possible

def main():
    # Load knowledge
    kb = load_knowledge_base()
    symptoms_list = kb["symptoms"]
    rules = kb["rules"]

    # Get user input
    user_symptoms = get_user_symptoms(symptoms_list)

    # Infer diseases
    diseases = infer_diseases(rules, user_symptoms)

    # Display result
    print("\n" + "="*40)
    if diseases:
        print("Possible illness(es):", ", ".join(diseases))
    else:
        print("No disease matches the reported symptoms.")
        print("Please consult a doctor for further evaluation.")
    print("="*40)

if __name__ == "__main__":
    main()
```
